```python
import statistics
```

```python
D = [10 , 20 , 30 , 40 , 50]
Mean = statistics.mean(D)
print(Mean)
```

```python
D = [5 , 15 , 25 , 35 , 45]
Median = statistics.median(D)
print(Median)
```

```python
D = [2 , 4 , 6 , 2 , 8 , 2 , 10]
Mode = statistics.mode(D)
print(Mode)
```

```python
D = [10 , 12 , 14 , 16 , 18]
Range = max(D) - min(D)
print(Range)
Variance = statistics.variance(D)
print(Variance)
```

```python
D = [3 , 6 , 9 , 12 , 15]
Standard_Deviation = statistics.stdev(D)
print(Standard_Deviation)
```
