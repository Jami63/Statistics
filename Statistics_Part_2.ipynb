```python
import matplotlib.pyplot as PLOT
```

```python
Marks = [45 , 50 , 55 , 60 , 65 , 70 , 72 , 75 , 80 , 85 , 90]
PLOT.hist(Marks , bins = 5)
PLOT.xlabel("Marks")
PLOT.ylabel("Frequency")
PLOT.title("Marks Distribution")
PLOT.show()
```

```python
import statistics
Scores = [40 , 45 , 50 , 55 , 60 , 65 , 70 , 75 , 80]
PLOT.boxplot(Scores)
PLOT.xlabel("Scores")
PLOT.ylabel("Frequency")
PLOT.title("Scores Distribution")
PLOT.show()
Median = statistics.median(Scores)
Q1 = statistics.quantiles(Scores)[0]
Q3 = statistics.quantiles(Scores)[2]
IQR = Q3 - Q1
print("Median : " , Median)
print("Q1 : " , Q1)
print("Q3 : " , Q3)
```

```python
Age = [18 , 19 , 20 , 21 , 22 , 23 , 24 , 60]
PLOT.boxplot(Age)
PLOT.xlabel("Age")
PLOT.ylabel("Frequency")
PLOT.title("Age Distribution")
PLOT.show()
Median = statistics.median(Age)
Q1 = statistics.quantiles(Age)[0]
Q3 = statistics.quantiles(Age)[2]
IQR = Q3 - Q1
print("Median : " , Median)
print("Q1 : " , Q1)
print("Q3 : " , Q3)
print("Outliers : " , [60])
```

```python
Class_A = [50 , 55 , 60 , 65 , 70 , 75]
Class_B = [45 , 50 , 55 , 60 , 85]
PLOT.boxplot([Class_A , Class_B])
PLOT.xlabel("Class")
PLOT.ylabel("Marks")
PLOT.title("Class Marks Distribution")
PLOT.show()
Median_A = statistics.median(Class_A)
Median_B = statistics.median(Class_B)
Q1_A = statistics.quantiles(Class_A)[0]
Q3_A = statistics.quantiles(Class_A)[2]
print("Median of Class A : " , Median_A)
print("Median of Class B : " , Median_B)
print("Q1 of Class A : " , Q1_A)
print("Q3 of Class A : " , Q3_A)
IQR_A = Q3_A - Q1_A
print("IQR of Class A : " , IQR_A)
IQR_B = statistics.quantiles(Class_B)[2] - statistics.quantiles(Class_B)[0]
print("IQR of Class B : " , IQR_B)
print("Median of Class A is greater than Class B")
```

```python
import numpy as N
Mu = 60
Sigma = 10
D = N.random.normal(Mu, Sigma, 100)
PLOT.hist(D , bins = 5)
PLOT.xlabel("Marks")
PLOT.ylabel("Frequency")
PLOT.title("Marks Distribution")
PLOT.show()
```

```python
from scipy.stats import norm
Mu = 1000
Sigma = 100
P1 = norm.cdf(1100, Mu, Sigma) - norm.cdf(900, Mu, Sigma)
P2 = norm.cdf(800 , Mu , Sigma)
print("P1 : " , P1 * 100 , "%")
print("P2 : " , P2 * 100 , "%")
X = N.linspace(700 , 1300 , 1000)
Y = norm.pdf(X , Mu , Sigma)
PLOT.plot(X , Y)
PLOT.xlabel("Bubble Life (H)")
PLOT.ylabel("Probability")
PLOT.title("Normal Distribution")
PLOT.show()
```
